```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import os

# File to Load (Remember to change these)
city_data_to_load = "Resource/city_data.csv"
ride_data_to_load = "Resource/ride_data.csv"

# Read the City and Ride Data
cityride_df=pd.read_csv(city_data_to_load)
ridedata_df=pd.read_csv(ride_data_to_load)

##Check for duplicates and drop data with same city name, same ride_ids but different # of drivers
#cityride_df = cityride_df.drop_duplicates('city', keep = 'first')

# Combine the data into a single dataset
##Merge and combine all data
merged_df = cityride_df.merge(ridedata_df, on = 'city')

# Display the data table for preview
merged_df.head()
```

```python
rural=merged_df[merged_df ["type"]=="Rural"]
suburban=merged_df[merged_df ["type"]=="Suburban"]
urban=merged_df[merged_df ["type"]=="Urban"]
#print (urban)
Urban_Ride_ID=urban.groupby(["city"]).count()["ride_id"]
Urban_Average_Fare=urban.groupby(["city"]).mean()["fare"]
Urban_Driver_Count=urban.groupby(["city"]).first()["driver_count"]

Suburban_Ride_ID=suburban.groupby(["city"]).count()["ride_id"]
Suburban_Average_Fare=suburban.groupby(["city"]).mean()["fare"]
Suburban_Driver_Count=suburban.groupby(["city"]).first()["driver_count"]

Rural_Ride_ID=rural.groupby(["city"]).count()["ride_id"]
Rural_Average_Fare=rural.groupby(["city"]).mean()["fare"]
Rural_Driver_Count=rural.groupby(["city"]).first()["driver_count"]
#print (Suburban_Ride_ID)
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types
plt.scatter(Urban_Ride_ID, Urban_Average_Fare, s=Urban_Driver_Count*10, 
            c="lightcoral", edgecolor = 'black', label="urban",alpha=.75)
plt.scatter(Suburban_Ride_ID, Suburban_Average_Fare, s=Suburban_Driver_Count*10, 
            c="lightskyblue", edgecolor = 'black', label="surburban", alpha=.75)
plt.scatter(Rural_Ride_ID, Rural_Average_Fare, s=Rural_Driver_Count*10, 
            c="gold", edgecolor = 'black', label="rural", alpha=.75)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")


# Create a legend
plt.legend(markerscale=.5, title="City Type")

# Incorporate a text label regarding circle size
plt.text(42,36," Note:\n Circle size correlates with driver count per city")
plt.grid()
# Save Figure
plt.savefig("figure1")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
#group_cityride_type=cityride_df.groupby(["type"])
#group_ridedata_type=ridedata_df.groupby(["type"])


group_type= merged_df.groupby('type')['type', 'fare', 'ride_id', 'driver_count']
totalfare= group_type.sum()['fare']

#***Build Pie Chart***

#colors
color_scheme = {'Gold':'#FFD700', 'Light Sky Blue':'#87CEFA', 'Light Coral':'#F08080'}
# sns.palplot(sns.color_palette(color_scheme.values()))
pie_labels = totalfare.index
city_color = {'Urban': color_scheme['Light Coral'], 'Suburban': color_scheme['Light Sky Blue'], 
              'Rural': color_scheme['Gold']}
colors = [city_color[n] for n in pie_labels]

explode = (0 , 0, .1)
plt.pie(totalfare, startangle = 150, colors = colors, explode = explode, labels = pie_labels, autopct = "%1.1f%%", 
        shadow = True, wedgeprops = {'linewidth': .5, 'edgecolor': 'black'})

# Incorporate the other graph properties
plt.title('% of Total Fares by City Type')

#explode = (0 , 0, .1)
plt.axis([-1.25,1.25,-1.25,1.25])

# Save Figure
plt.savefig("figure2")
#plt.tight_layout()
plt.show()
```
